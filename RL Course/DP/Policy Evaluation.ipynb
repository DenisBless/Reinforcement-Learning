# Policy evaluation
In policy evaluation we are given a policy (Function that assigns probabilty for actions depending on the state $\pi(a|s)$). As $k$ approaches infinity the value function $v_k$ converges to the best value function $v_\pi$ under policy $\pi$. However, the value function $v_\pi$ is not necessarily the optimal value function $v_*$ since we followed an arbitrary policy. The pseudocode for the algorithm is shown below.

```python
def policy_eval(policy, env, discount_factor=1.0, theta=0.00001):
    """
    Evaluate a policy given an environment and a full description of the environment's dynamics.
    
    Args:
        policy: [S, A] shaped matrix representing the policy.
        env: OpenAI env. env.P represents the transition probabilities of the environment.
            env.P[s][a] is a list of transition tuples (prob, next_state, reward, done).
            env.nS is a number of states in the environment. 
            env.nA is a number of actions in the environment.
        theta: We stop evaluation once our value function change is less than theta for all states.
        discount_factor: Gamma discount factor.
    
    Returns:
        Vector of length env.nS representing the value function.
    """
    # Start with a random (all 0) value function
    V = np.zeros(env.nS).reshape(-1,1)
    state_transition = np.zeros((env.nS, env.nS))
    rewards = np.zeros(env.nS)
    for s in range(env.nS):
        # Rewards
        rewards[s] = np.sum([(policy[s][a] * env.P[s][a][0][2]) for a in range(env.nA)])
        # Indices for state transition matrix
        next_states = np.asarray([env.P[s][a][0][1] for a in range(env.nA)])
        # State transistion matrix
        for i, s_prime in enumerate(next_states):
            state_transition[s,s_prime] += random_policy[s,i] * 1
    
    while True:
        v = V
        # Vectorized solution
        V = rewards.reshape(-1,1) + np.dot(state_transition,V).reshape(-1,1)
        delta = np.abs(v-V)
        # Break if smaller threshold
        if delta.any() < theta:
            break

    return np.array(V)
```

```python
random_policy = np.ones([env.nS, env.nA]) / env.nA #  Assigns 1/n pr. to the n actions
v = policy_eval(random_policy, env)
print('final value fn:', *v)
```
